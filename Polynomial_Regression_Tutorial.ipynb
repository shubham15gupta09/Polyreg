#Polynominal Regression Model
---

\\

The model of polynominal Regression is 

\\

$$y_1 = \beta_0 + \beta_1x_i + \beta_2x^2_i + \cdots + \beta_mx^m_i + \varepsilon_i\ (i = 1,2,\cdots , n) $$

\\

and this model can be expressed in martix form in terms of a desgin martix $\mathbf{X}$ which is a response vector $\vec y$, a parameter vector $\vec \beta$ and a vector $\vec\varepsilon$ of random errors. The $i^{th}$ row of $\mathbf{X}$ and $\vec y$ will contain the $x$ and $y$ value for the $i$th data sample. Then the model can be written as a system of linear equations

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('data.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values

# Fitting Linear Regression to the dataset 
from sklearn.linear_model import LinearRegression 
lin = LinearRegression() 
  
lin.fit(X, y)

lin2 = LinearRegression()

# Visualising the Polynomial Regression results 
plt.scatter(X, y, color = 'blue') 
  
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Salary') 
plt.ylabel('Position') 
  
plt.show() 

 
# Predicting a new result with Linear Regression 
lin.predict(110.0)

lin2.predict(poly.fit_transform(110.0))

````
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv("https://drive.google.com/file/d/1tNL4jxZEfpaP4oflfSn6pIHJX7Pachm9/view")
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values


# Fitting Linear Regression to the dataset 
from sklearn.linear_model import LinearRegression 
lin = LinearRegression() 
  
lin.fit(X, y) 

# Fitting Polynomial Regression to the dataset 
from sklearn.preprocessing import PolynomialFeatures 
  
poly = PolynomialFeatures(degree = 4) 
X_poly = poly.fit_transform(X) 
  
poly.fit(X_poly, y) 
lin2 = LinearRegression() 
lin2.fit(X_poly, y) 
 
# Visualising the Linear Regression results 
plt.scatter(X, y, color = 'blue') 
  
plt.plot(X, lin.predict(X), color = 'red') 
plt.title('Linear Regression') 
plt.xlabel('Salary') 
plt.ylabel('Position') 
  
plt.show() 

 
# Visualising the Polynomial Regression results 
plt.scatter(X, y, color = 'blue') 
  
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Salary') 
plt.ylabel('Position') 
  
plt.show() 

 
# Predicting a new result with Linear Regression 
lin.predict(110.0) 


# Predicting a new result with Polynomial Regression 
lin2.predict(poly.fit_transform(110.0)) 
````
